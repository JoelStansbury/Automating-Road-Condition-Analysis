```python
import numpy as np
import numpy.random as nprd
import numpy.linalg as npla
import matplotlib.pyplot as plt
import os
import pickle
from tqdm import tqdm
import scipy.signal
import csv

#matrix with: latitue, longitude, degree, cluster
#go through these lists:
#put the first point into the first key of the dictionary, 0, with value which is a list of lists: 
#[[list of lats and longs(lat1, long1), (lat2, long2), ...], [list of cluster assignments c1, c2, ...], 
#[list of degrees d1, d2, ...]
#take this point off of points to go through
#go through the next point, if latitude is within .1 of previous latitude, append it to values of 0
#if not, continue until the next point that is, keep going until end of list
#next dictionary key 1 with the first point that didn't fall into 0 and all the other points that are within .1 of it
#keep going until no more
#should result in about 5-20
#if > 20 --> needs to be a smaller difference than .1, if less than 5, greater

def creatematrix(filename):
    matrix = np.genfromtxt(filename, delimiter=',', skip_header=1)
    return matrix
            

def createdict(matrix):
    msize = np.size(matrix, 0)
#     matrix = np.array([matrix,]*msize)
    dictionary = {}
    keycount = 0
    for i in range(msize):
        latlong = (matrix[i][0], matrix[i][1])
        degree = matrix[i][2]
        cluster = matrix[i][3]
        values = [[latlong], [degree], [cluster]]
#         print("Row (an anomaly):", dict_values)
                
        if i == 0:
            # Create new entry for first anomaly
            dictionary[keycount] = values 
        elif latlong not in dictionary.values():
            # If anomaly not already in dictionary, create a new entry for it
            dictionary.update({keycount: values})
        else: 
            continue
            
        keycount += 1
        next_anom = i+1
        while next_anom < msize:    
            next_latlong = (matrix[next_anom][0], matrix[next_anom][1])
            next_degree = matrix[next_anom][2]
            next_cluster = matrix[next_anom][3]
            if next_latlong not in dictionary.values():
                curr_long = matrix[i][1]
                next_long = matrix[next_anom][1]
                if np.all([next_long <= curr_long + .000025, next_long >= curr_long - .000025]):
                    values[0].append(next_latlong)
                    values[1].append(next_degree)
                    values[2].append(next_cluster)
                next_anom += 1

    return dictionary
            
result_m = creatematrix("output.csv")
result_d = createdict(result_m)
for keys, values in result_d.items():
    print(keys)
    print(values)
    
    
# range for first anomaly: [-83.92723487, -83.92695987]
    
```
